## Section 1. Statistical Test

#### 1.1 Which statistical test did you use to analyze the NYC subway data? Did you use a one-tail or a two-tail P value? What is the null hypothesis? What is your p-critical value?

I used  Mann-Whitney U-Test one-tail test. P-critical value is 0.05 (95% of confidence). Null hypothesis is ridership in NYC Subway is the same in rainy and non-rainy days.

#### 1.2 Why is this statistical test applicable to the dataset?

We couldn’t use Welch's t-Test because our data was not normally distributed but skewed.

#### 1.3 What results did you get from this statistical test? These should include the following numerical values: p-values, as well as the means for each of the two samples under test. What is the significance and interpretation of these results?

 Our p-value == 0.025 is less than 0.05 (5%), which means that we reject the Null Hypothesis. Using Mann-Whitney U-Test we have found two mean values: 1105.45 is the mean of entries with rain and 1090.28 is the mean of entries without rain. According to the test, we found that these two result are significantly different. So we can state that more people use metro in rainy weather.
